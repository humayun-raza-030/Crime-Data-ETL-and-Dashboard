# Data Cleaning & Transformations


```python
df = df.drop(columns=['Unnamed: 0'])
```

```python
# Check missing values
print(f"Missing Latitude: {df['Latitude'].isnull().sum()}")
print(f"Missing Longitude: {df['Longitude'].isnull().sum()}")
```

```python
# Drop rows with missing latitude or longitude
df = df.dropna(subset=['Latitude', 'Longitude'])
```

```python
# Convert to datetime
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y %I:%M:%S %p', errors='coerce')

# Drop rows where Date conversion failed
df = df.dropna(subset=['Date'])

# Normalize timestamps into components
df['Year_Extracted'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df['Hour'] = df['Date'].dt.hour
df['Weekday'] = df['Date'].dt.day_name()
```

```python
df.head()
```

```python
# Check duplicates
print(f"Duplicate IDs: {df.duplicated(subset=['ID']).sum()}")

# Drop duplicates based on ID
df = df.drop_duplicates(subset=['ID'])
```

```python
# List of columns to uppercase and strip spaces
textual_columns = ['Case Number', 'Block', 'IUCR', 'Primary Type', 
                   'Description', 'Location Description', 'FBI Code']

# Apply upper and strip
for col in textual_columns:
    if col in df.columns:
        df[col] = df[col].astype(str).str.upper().str.strip()

# For 'Arrest' and 'Domestic', ensure proper boolean type
bool_columns = ['Arrest', 'Domestic']
for col in bool_columns:
    if col in df.columns:
        df[col] = df[col].astype(bool)

# For Ward and Community Area, fill missing and cast to integer if needed
ward_columns = ['Ward', 'Community Area']
for col in ward_columns:
    if col in df.columns:
        df[col] = df[col].fillna(-1).astype(int)
```

| Column | Action |
| --- | --- |
| Case Number | Uppercased, stripped spaces |
| Block | Uppercased, stripped spaces |
| IUCR | Uppercased, stripped spaces |
| Primary Type | Uppercased, stripped spaces |
| Description | Uppercased, stripped spaces |
| Location Description | Uppercased, stripped spaces |
| FBI Code | Uppercased, stripped spaces |
| Arrest | Ensured Boolean values |
| Domestic | Ensured Boolean values |
| Ward | Filled missing with -1, converted to int |
| Community Area | Filled missing with -1, converted to int |

# Database Schema (Normalization)

```python
# Extract Crime Types
crime_types = df[['Primary Type', 'Description']].drop_duplicates().reset_index(drop=True)

# Create unique Crime_Type_ID
crime_types['Crime_Type_ID'] = crime_types.index + 1

# Reorder columns
crime_types = crime_types[['Crime_Type_ID', 'Primary Type', 'Description']]
crime_types.head()
```

```python
# Extract Locations
locations = df[['Block', 'Location Description', 'Latitude', 'Longitude', 'Ward', 'Community Area']].drop_duplicates().reset_index(drop=True)

# Create unique Location_ID
locations['Location_ID'] = locations.index + 1

# Reorder columns
locations = locations[['Location_ID', 'Block', 'Location Description', 'Latitude', 'Longitude', 'Ward', 'Community Area']]
locations.head()
```

```python
# Merge Crime_Types to get Crime_Type_ID
df = df.merge(crime_types, on=['Primary Type', 'Description'], how='left')

# Merge Locations to get Location_ID
df = df.merge(locations, on=['Block', 'Location Description', 'Latitude', 'Longitude', 'Ward', 'Community Area'], how='left')

# Create Incidents table
incidents = df[['ID', 'Case Number', 'Date', 'Arrest', 'Domestic', 
                'Crime_Type_ID', 'Location_ID', 'IUCR', 'FBI Code', 
                'Year_Extracted', 'Month', 'Day', 'Hour', 'Weekday']]

# Rename columns nicely
incidents.rename(columns={'ID': 'Incident_ID', 
                          'Case Number': 'Case_Number', 
                          'FBI Code': 'FBI_Code', 
                          'Year_Extracted': 'Year'}, inplace=True)
```

# Feature Engineering & Table Reshaping

## Spatial Density (Incidents per km² per Community Area)

```python
print(f"Average Community Area: {df['Community Area'].mean()}")
```

```python
# Total number of incidents per Community Area
community_counts = df.groupby('Community Area').size().reset_index(name='Total_Incidents')

# Assume average community area size
avg_area_km2 = df['Community Area'].mean()

# Calculate spatial density
community_counts['Spatial_Density'] = community_counts['Total_Incidents'] / avg_area_km2

# Merge back to Locations
locations = locations.merge(community_counts[['Community Area', 'Spatial_Density']], on='Community Area', how='left')
```

```python
locations.head()
```

## Rolling 7-day Average of Incident Counts

```python
# Convert 'Date' to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Sort by Date
df = df.sort_values('Date')

# Create a daily incident count
daily_counts = df.groupby(df['Date'].dt.date).size().reset_index(name='Daily_Incidents')

# Rolling 7-day average
daily_counts['Rolling_7D_Avg'] = daily_counts['Daily_Incidents'].rolling(window=7, min_periods=1).mean()

# Merge rolling average back to incidents
df = df.merge(daily_counts[['Date', 'Rolling_7D_Avg']], left_on=df['Date'].dt.date, right_on='Date', how='left')
df
```

## Weekend/Weekday Flag

```python
# Create a flag: 1 if Weekend, 0 if Weekday
df['Weekend_Flag'] = df['Date_x'].dt.weekday.apply(lambda x: 1 if x >= 5 else 0)
df
```

## Repeat Incident Probability

```python
# Count incidents per Block
block_counts = df.groupby('Block').size().reset_index(name='Block_Incidents')

# Total incidents
total_incidents = len(df)

# Probability that a random incident is from a frequently-hit block
block_counts['Repeat_Incident_Prob'] = block_counts['Block_Incidents'] / total_incidents

# Merge back to Locations
locations = locations.merge(block_counts[['Block', 'Repeat_Incident_Prob']], on='Block', how='left')
locations
```

## Season of Crime

```python
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

df['Crime_Season'] = df['Month'].apply(get_season)
df
```

## Crime Severity Score

```python
# Get unique values from 'Primary Type' column
primary_types = df['Primary Type'].unique()
print(f"Number of unique crime types: {len(primary_types)}")
print("\nUnique crime types:")
for crime_type in sorted(primary_types):
    print(f"- {crime_type}")
```

```python
# Define severity mapping
severity_mapping = {
    'ARSON': 4,
    'ASSAULT': 4,
    'BATTERY': 4,
    'BURGLARY': 3,
    'CONCEALED CARRY LICENSE VIOLATION': 2,
    'CRIM SEXUAL ASSAULT': 5,
    'CRIMINAL DAMAGE': 3,
    'CRIMINAL TRESPASS': 1,
    'DECEPTIVE PRACTICE': 3,
    'GAMBLING': 2,
    'HOMICIDE': 5,
    'HUMAN TRAFFICKING': 5,
    'INTERFERENCE WITH PUBLIC OFFICER': 2,
    'INTIMIDATION': 1,
    'KIDNAPPING': 5,
    'LIQUOR LAW VIOLATION': 2,
    'MOTOR VEHICLE THEFT': 3,
    'NARCOTICS': 3,
    'NON - CRIMINAL': 1,
    'NON-CRIMINAL': 1,
    'NON-CRIMINAL (SUBJECT SPECIFIED)': 1,
    'OBSCENITY': 2,
    'OFFENSE INVOLVING CHILDREN': 1,
    'OTHER NARCOTIC VIOLATION': 2,
    'OTHER OFFENSE': 2,
    'PROSTITUTION': 2,
    'PUBLIC INDECENCY': 1,
    'PUBLIC PEACE VIOLATION': 2,
    'ROBBERY': 4,
    'SEX OFFENSE': 1,
    'STALKING': 1,
    'THEFT': 3,
    'WEAPONS VIOLATION': 4
}


# Apply mapping
df['Crime_Severity_Score'] = df['Primary Type'].map(severity_mapping)

# Fill missing with 1 (default severity)
df['Crime_Severity_Score'] = df['Crime_Severity_Score'].fillna(1)
```

##  Unpivot Crime Counts

```python
# Group by Year, Month, Primary Type
crime_counts = df.groupby(['Year', 'Month', 'Primary Type']).size().reset_index(name='Crime_Count')

# This is already long-format (unpivoted).
crime_counts
```

## Pivot monhtly

```python
# Pivot so that each crime type becomes a column
crime_pivot = crime_counts.pivot_table(
    index=['Year', 'Month'],
    columns='Primary Type',
    values='Crime_Count',
    fill_value=0
)

# Reset index if needed for easy access
crime_pivot = crime_pivot.reset_index()

crime_pivot
```

| Step | Action | Output |
| --- | --- | --- |
| 1 | Group by Year, Month, Primary_Type | Unpivoted long table |
| 2 | Pivot for time series | Wide table with crime types as columns |
